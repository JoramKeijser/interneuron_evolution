```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import anndata as ad
import scanpy as sc
sns.set_context("poster")
sns.set_palette("Set2")
```

```python
gaba = ad.read_h5ad("../data/raw/mouse_bird_gaba_integrated.h5ad")
gaba
```

```python
fig, ax = plt.subplots()
sc.pl.umap(gaba, color='organism', frameon=False, title='GABAergic neurons', ax=ax)#save="_integrated_gaba")
fig.savefig("../figures/integrated_gaba", dpi=300)
```
